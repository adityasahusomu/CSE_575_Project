```python
def crop_and_resize_batch(image_batch, label_batch, add_pixels_value=0):
    """
    Processes a batch of images for cropping and resizing, keeping labels unchanged.
    """
    processed_images = []
    for image in image_batch:
        # Convert tensor to NumPy array
        img = tf.keras.preprocessing.image.img_to_array(image)
        img = img.astype(np.uint8)

        # Convert to grayscale and apply OpenCV processing
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        gray = cv2.GaussianBlur(gray, (5, 5), 0)

        # Threshold and find contours
        thresh = cv2.threshold(gray, 45, 255, cv2.THRESH_BINARY)[1]
        thresh = cv2.erode(thresh, None, iterations=2)
        thresh = cv2.dilate(thresh, None, iterations=2)
        cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = imutils.grab_contours(cnts)

        if len(cnts) > 0:
            # Get the largest contour and extreme points
            c = max(cnts, key=cv2.contourArea)
            extLeft = tuple(c[c[:, :, 0].argmin()][0])
            extRight = tuple(c[c[:, :, 0].argmax()][0])
            extTop = tuple(c[c[:, :, 1].argmin()][0])
            extBot = tuple(c[c[:, :, 1].argmax()][0])

            # Crop and pad
            ADD_PIXELS = add_pixels_value
            new_img = img[extTop[1]-ADD_PIXELS:extBot[1]+ADD_PIXELS,
                          extLeft[0]-ADD_PIXELS:extRight[0]+ADD_PIXELS].copy()
        else:
            # If no contours are found, use the original image
            new_img = img

        # Resize the cropped image back to the desired size (224x224)
        new_img = cv2.resize(new_img, (224, 224), interpolation=cv2.INTER_CUBIC)
        processed_images.append(new_img)

    # Normalize and convert back to TensorFlow tensor
    processed_images = tf.convert_to_tensor(processed_images, dtype=tf.float32) / 255.0
    # Ensure labels have the correct shape
    return processed_images, tf.convert_to_tensor(label_batch, dtype=tf.int32)
```

```python
# Create datasets
train_ds_custom = keras.utils.image_dataset_from_directory(directory=train_dir,
                                                           labels='inferred',
                                                           label_mode='int',
                                                           batch_size=16,
                                                           image_size=(224, 224))

validation_ds_custom = keras.utils.image_dataset_from_directory(directory=test_dir,
                                                                labels='inferred',
                                                                label_mode='int',
                                                                batch_size=16,
                                                                image_size=(224, 224))

# Define the shape-setting function
def set_shapes(image, label):
    image.set_shape((None, 224, 224, 3))
    label.set_shape((None,))
    return image, label

# Map preprocessing function and set shapes
train_ds_custom = train_ds_custom.map(
    lambda x, y: set_shapes(*tf.py_function(func=crop_and_resize_batch, inp=[x, y], Tout=(tf.float32, tf.int32)))
)

validation_ds_custom = validation_ds_custom.map(
    lambda x, y: set_shapes(*tf.py_function(func=crop_and_resize_batch, inp=[x, y], Tout=(tf.float32, tf.int32)))
)

# Prefetch for performance
train_ds_custom = train_ds_custom.prefetch(buffer_size=tf.data.AUTOTUNE)
validation_ds_custom = validation_ds_custom.prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
# Custom CNN Architecture:

model = Sequential()

model.add(Conv2D(1024, kernel_size = (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides = 2))

model.add(Conv2D(512, kernel_size = (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides = 2))

model.add(Conv2D(256, kernel_size = (3, 3), activation='relu'))
model.add(Conv2D(128, kernel_size = (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides = 2))

model.add(Conv2D(64, kernel_size = (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides = 2))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(4, activation='softmax'))
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
import os

local_save_path = '/content/BEST_custom.keras'

os.makedirs(os.path.dirname(local_save_path), exist_ok=True)

checkpoint = ModelCheckpoint(filepath=local_save_path,
                              monitor='val_loss',
                              save_best_only=True,
                              save_weights_only=False,
                              verbose=1)

reduce_learning_rate = ReduceLROnPlateau(monitor='val_loss',
                                          factor=0.1,
                                          patience=6,
                                          verbose=1)

early_stopping = EarlyStopping(monitor='val_accuracy',
                               verbose=1,
                               patience=10)

callbacks = [checkpoint, reduce_learning_rate, early_stopping]
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model('/content/BEST_custom.keras')

predict_list = []
true_labels = []

for batch in validation_ds_custom:

    images, labels = batch
    true_labels.extend(labels.numpy())

    for img in images:
        img_array = np.array(img)
        img_array = img_array.reshape(1, 224, 224, 3)

        pred = model.predict(img_array)

        indices = pred.argmax()

        predict_list.append(indices)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

labels = ['glioma', 'meningioma', 'notumor', 'pituitary']
cm = confusion_matrix(true_labels, predict_list)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
fig, ax = plt.subplots(figsize=(8, 8))
disp.plot(ax=ax, cmap='Blues', xticks_rotation='vertical')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(true_labels, predict_list))
```
